```python
from pymongo import MongoClient
```

```python
# Поключение к MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["W3T2_db"]
collect = db["storage_books"]
```

```python
# Запрос количества и названий книг, дороже 50 фунтов
query = {"Price": {"$gt": 56.00}}

# Получение документов, соответствующих запросу
documents = collect.find(query)

# Подсчет количества документов, соответствующих запросу
count = collect.count_documents(query)

print("Количество книг дороже 56 фунтов - ", count, ":")

# Отображение названий книг
for document in documents:
    print("Название книги:", document['Title'])
```

```python
# Запрос для определения максимальной и минимальной цены книги в коллекции
# Определение конвеера для агрегации данных
pipeline = [
    {"$group": {"_id": None, "max_price": {"$max": "$Price"}, "min_price": {"$min": "$Price"}}}
]

# Применение конвеера
result = list(collect.aggregate(pipeline))

# Получение минимального и максимального значений
max_price = result[0]["max_price"]
min_price = result[0]["min_price"]

# Печать результатов
print("Минимальная цена:", min_price)
print("Максимальная цена:", max_price)
```
